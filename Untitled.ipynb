```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Button, Slider
from mpl_toolkits.mplot3d import Axes3D
from scipy.ndimage import distance_transform_edt
import matplotlib
import speech_recognition as sr
from gtts import gTTS
import os
import platform
import subprocess

# Set the backend to TkAgg for interactive features
matplotlib.use('TkAgg')

# Function for TTS
def speak_message(message):
    tts = gTTS(text=message, lang='en')  # Convert text to speech
    tts.save("message.mp3")  # Save the speech as an MP3 file

    # Detect OS and play sound
    if platform.system() == "Darwin":  # macOS
        subprocess.run(["afplay", "message.mp3"])
    elif platform.system() == "Linux":  # Linux
        subprocess.run(["xdg-open", "message.mp3"])
    elif platform.system() == "Windows":  # Windows
        os.system("start message.mp3")
    else:
        print("Unsupported OS")

# Create a blank 2D canvas (white canvas: 1 = white, 0 = black)
canvas = np.ones((500, 500))

# Create a Matplotlib figure and axis for the 2D canvas
fig, (ax2d, ax3d) = plt.subplots(1, 2, figsize=(10, 5), subplot_kw={"projection": None})
im = ax2d.imshow(canvas, cmap="gray", vmin=0, vmax=1)
ax2d.axis("off")  # Hide axes for a cleaner 2D canvas

# Initialize variables for drawing and erasing
drawing = False
brush_size = 5

# Function to convert 2D canvas into a 3D model
def update_3d():
    ax3d.clear()
    ax3d.set_title("3D Model")
    ax3d.axis("off")

    # Use distance transform to create heightmap
    inverted_canvas = 1 - canvas
    heightmap = distance_transform_edt(inverted_canvas)

    # Create 3D surface
    x, y = np.meshgrid(np.arange(canvas.shape[1]), np.arange(canvas.shape[0]))
    ax3d.plot_surface(
        x, y, heightmap, cmap="viridis", edgecolor="none", alpha=0.8
    )
    fig.canvas.draw_idle()

# Function to draw a circle on the canvas
def draw_circle(x, y, radius, color=0):
    global canvas
    x, y = int(x), int(y)
    for i in range(-radius, radius + 1):
        for j in range(-radius, radius + 1):
            if 0 <= x + i < canvas.shape[1] and 0 <= y + j < canvas.shape[0]:
                if i**2 + j**2 <= radius**2:  # Check for circular shape
                    canvas[y + j, x + i] = color
    im.set_data(canvas)
    update_3d()

# Function to draw a rectangle on the canvas
def draw_rectangle(x, y, width, height, color=0):
    global canvas
    x, y = int(x), int(y)
    for i in range(-width // 2, width // 2 + 1):
        for j in range(-height // 2, height // 2 + 1):
            if 0 <= x + i < canvas.shape[1] and 0 <= y + j < canvas.shape[0]:
                canvas[y + j, x + i] = color
    im.set_data(canvas)
    update_3d()

# Voice command handling
def handle_voice_command(command):
    """Executes drawing actions based on voice commands."""
    if "circle" in command:
        draw_circle(canvas.shape[1] // 2, canvas.shape[0] // 2, brush_size * 5)
        speak_message("Drawing a circle.")
    elif "rectangle" in command:
        draw_rectangle(canvas.shape[1] // 2, canvas.shape[0] // 2, brush_size * 10, brush_size * 5)
        speak_message("Drawing a rectangle.")
    elif "clear" in command:
        clear_canvas(None)
        speak_message("Canvas cleared.")
    else:
        speak_message("Command not recognized. Please say 'Draw a circle,' 'Draw a rectangle,' or 'Clear.'")
        print("Command not recognized.")

# Function to process voice input
def process_voice_input(event):
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for a command...")
        speak_message("I'm listening. Please say a command.")
        try:
            audio = recognizer.listen(source, timeout=5)
            command = recognizer.recognize_google(audio).lower()
            print(f"Command received: {command}")
            handle_voice_command(command)
        except sr.UnknownValueError:
            speak_message("Sorry, I couldn't understand the command.")
            print("Could not understand the command.")
        except sr.RequestError as e:
            speak_message("There was an error with the speech recognition service.")
            print(f"Could not request results; {e}")
        except sr.WaitTimeoutError:
            speak_message("I did not hear anything. Please try again.")
            print("Listening timed out.")

# Clear canvas function
def clear_canvas(event):
    global canvas
    canvas.fill(1)  # Reset canvas to white
    im.set_data(canvas)
    update_3d()

# Add a "Clear" button to the figure
clear_ax = plt.axes([0.8, 0.05, 0.1, 0.075])  # Button position
clear_button = Button(clear_ax, "Clear")
clear_button.on_clicked(clear_canvas)

# Slider to adjust brush size
brush_slider_ax = plt.axes([0.2, 0.01, 0.4, 0.03])  # Slider position
brush_slider = Slider(brush_slider_ax, "Brush Size", 1, 20, valinit=brush_size, valstep=1)

def update_brush_size(val):
    global brush_size
    brush_size = int(val)

brush_slider.on_changed(update_brush_size)

# Add a "Voice Input" button to the figure
voice_ax = plt.axes([0.65, 0.05, 0.1, 0.075])  # Button position
voice_button = Button(voice_ax, "Voice Input")
voice_button.on_clicked(process_voice_input)

# Add a subplot for 3D modeling
ax3d = fig.add_subplot(122, projection='3d')
update_3d()  # Initialize the 3D view

# Show the 2D and 3D canvas
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Button, Slider
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from scipy.interpolate import splprep, splev

# Initialize figure and subplots
fig = plt.figure(figsize=(10, 6))
ax2d = fig.add_subplot(121)
ax3d = fig.add_subplot(122, projection='3d')

ax2d.set_title("Draw Here (2D Canvas)")
ax3d.set_title("3D Model (Extruded Shape)")

# Store stroke points and settings
stroke_points = []
drawing = False
brush_size = 2
draw_mode = "freehand"  # Modes: freehand, circle, rectangle

# Function to smooth strokes using splines
def smooth_stroke(points, num_interp=100):
    if len(points) < 4:
        return points  # Not enough points to interpolate

    x, y = zip(*points)
    tck, u = splprep([x, y], s=2)
    u_new = np.linspace(0, 1, num_interp)
    smooth_x, smooth_y = splev(u_new, tck)
    return list(zip(smooth_x, smooth_y))

# Function to extrude a shape into 3D
def extrude_shape(points, depth=20):
    if len(points) < 2:
        return

    points = smooth_stroke(points)  # Apply smoothing

    x_vals = [p[0] for p in points]
    y_vals = [p[1] for p in points]

    for z_offset in [0, depth]:
        ax3d.plot(x_vals, y_vals, [z_offset] * len(x_vals), 'k')

    for i in range(len(points) - 1):
        ax3d.plot([x_vals[i], x_vals[i]], [y_vals[i], y_vals[i]], [0, depth], 'k')

# Update the 3D model
def draw_3d_model():
    ax3d.clear()
    ax3d.set_title("3D Model (Extruded Shape)")
    ax3d.axis("off")

    if len(stroke_points) > 1:
        extrude_shape(stroke_points)

    fig.canvas.draw_idle()

# Handle mouse press event
def on_press(event):
    global drawing, stroke_points
    if event.inaxes == ax2d:
        drawing = True
        stroke_points.clear()

        if draw_mode == "circle":
            draw_circle(event.xdata, event.ydata, 20)
        elif draw_mode == "rectangle":
            draw_rectangle(event.xdata, event.ydata, 40, 20)
        else:
            stroke_points.append((event.xdata, event.ydata))

# Handle mouse movement for continuous drawing
def on_motion(event):
    if drawing and event.inaxes == ax2d and draw_mode == "freehand":
        stroke_points.append((event.xdata, event.ydata))
        ax2d.plot(event.xdata, event.ydata, 'ko', markersize=brush_size)
        fig.canvas.draw()

# Handle mouse release
def on_release(event):
    global drawing
    drawing = False
    draw_3d_model()

# Draw a circle
def draw_circle(cx, cy, radius):
    angles = np.linspace(0, 2*np.pi, 30)
    circle = [(cx + radius * np.cos(a), cy + radius * np.sin(a)) for a in angles]
    stroke_points.extend(circle)
    ax2d.plot(*zip(*circle), 'k-')
    fig.canvas.draw()
    draw_3d_model()

# Draw a rectangle
def draw_rectangle(cx, cy, width, height):
    rect = [(cx - width / 2, cy - height / 2), (cx + width / 2, cy - height / 2),
            (cx + width / 2, cy + height / 2), (cx - width / 2, cy + height / 2), (cx - width / 2, cy - height / 2)]
    stroke_points.extend(rect)
    ax2d.plot(*zip(*rect), 'k-')
    fig.canvas.draw()
    draw_3d_model()

# Clear canvas
def clear_canvas(event):
    global stroke_points
    stroke_points = []
    ax2d.clear()
    ax3d.clear()
    ax2d.set_title("Draw Here (2D Canvas)")
    ax3d.set_title("3D Model (Extruded Shape)")
    fig.canvas.draw()

# Change brush size
def update_brush_size(val):
    global brush_size
    brush_size = int(val)

# Toggle shape mode
def toggle_shape(event):
    global draw_mode
    modes = ["freehand", "circle", "rectangle"]
    draw_mode = modes[(modes.index(draw_mode) + 1) % len(modes)]
    shape_button.label.set_text(f"Mode: {draw_mode.capitalize()}")

# Add buttons and slider
ax_clear = plt.axes([0.3, 0.01, 0.1, 0.05])
clear_button = Button(ax_clear, "Clear")
clear_button.on_clicked(clear_canvas)

ax_shape = plt.axes([0.42, 0.01, 0.15, 0.05])
shape_button = Button(ax_shape, "Mode: Freehand")
shape_button.on_clicked(toggle_shape)

ax_slider = plt.axes([0.6, 0.01, 0.3, 0.03])
brush_slider = Slider(ax_slider, "Brush Size", 1, 10, valinit=2)
brush_slider.on_changed(update_brush_size)

# Connect events
fig.canvas.mpl_connect('button_press_event', on_press)
fig.canvas.mpl_connect('motion_notify_event', on_motion)
fig.canvas.mpl_connect('button_release_event', on_release)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.widgets import Button, Slider
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from scipy.interpolate import splprep, splev
import matplotlib
matplotlib.use('TkAgg')

# Adjust DPI based on screen size
mpl.rcParams['figure.dpi'] = 100  # Adjust for better scaling on different screens

# Function to dynamically set figure size
def get_figure_size():
    import platform
    if platform.system() == "Darwin":  # MacBook
        return (12, 7)
    elif platform.system() == "Windows":  # Windows
        return (10, 6)
    else:  # Default (Tablets, Phones)
        return (8, 5)

# Initialize figure and subplots
fig = plt.figure(figsize=get_figure_size())
ax2d = fig.add_subplot(121)
ax3d = fig.add_subplot(122, projection='3d')

ax2d.set_title("Draw Here (2D Canvas)", fontsize=12)
ax3d.set_title("3D Model (Extruded Shape)", fontsize=12)

# Store stroke points and settings
stroke_points = []
drawing = False
brush_size = 2
draw_mode = "freehand"  # Modes: freehand, circle, rectangle

# Function to smooth strokes using splines
def smooth_stroke(points, num_interp=100):
    if len(points) < 4:
        return points  # Not enough points to interpolate

    x, y = zip(*points)
    tck, u = splprep([x, y], s=2)
    u_new = np.linspace(0, 1, num_interp)
    smooth_x, smooth_y = splev(u_new, tck)
    return list(zip(smooth_x, smooth_y))

# Function to extrude a shape into 3D
def extrude_shape(points, depth=20):
    if len(points) < 2:
        return

    points = smooth_stroke(points)  # Apply smoothing

    x_vals = [p[0] for p in points]
    y_vals = [p[1] for p in points]

    for z_offset in [0, depth]:
        ax3d.plot(x_vals, y_vals, [z_offset] * len(x_vals), 'k')

    for i in range(len(points) - 1):
        ax3d.plot([x_vals[i], x_vals[i]], [y_vals[i], y_vals[i]], [0, depth], 'k')

# Update the 3D model
def draw_3d_model():
    ax3d.clear()
    ax3d.set_title("3D Model (Extruded Shape)", fontsize=12)
    ax3d.axis("off")

    if len(stroke_points) > 1:
        extrude_shape(stroke_points)

    fig.canvas.draw_idle()

# Handle mouse press event
def on_press(event):
    global drawing, stroke_points
    if event.inaxes == ax2d:
        drawing = True
        stroke_points.clear()

        if draw_mode == "circle":
            draw_circle(event.xdata, event.ydata, 20)
        elif draw_mode == "rectangle":
            draw_rectangle(event.xdata, event.ydata, 40, 20)
        else:
            stroke_points.append((event.xdata, event.ydata))

# Handle mouse movement for continuous drawing
def on_motion(event):
    if drawing and event.inaxes == ax2d and draw_mode == "freehand":
        stroke_points.append((event.xdata, event.ydata))
        ax2d.plot(event.xdata, event.ydata, 'ko', markersize=brush_size)
        fig.canvas.draw()

# Handle mouse release
def on_release(event):
    global drawing
    drawing = False
    draw_3d_model()

# Draw a circle
def draw_circle(cx, cy, radius):
    angles = np.linspace(0, 2*np.pi, 30)
    circle = [(cx + radius * np.cos(a), cy + radius * np.sin(a)) for a in angles]
    stroke_points.extend(circle)
    ax2d.plot(*zip(*circle), 'k-')
    fig.canvas.draw()
    draw_3d_model()

# Draw a rectangle
def draw_rectangle(cx, cy, width, height):
    rect = [(cx - width / 2, cy - height / 2), (cx + width / 2, cy - height / 2),
            (cx + width / 2, cy + height / 2), (cx - width / 2, cy + height / 2), (cx - width / 2, cy - height / 2)]
    stroke_points.extend(rect)
    ax2d.plot(*zip(*rect), 'k-')
    fig.canvas.draw()
    draw_3d_model()

# Clear canvas
def clear_canvas(event):
    global stroke_points
    stroke_points = []
    ax2d.clear()
    ax3d.clear()
    ax2d.set_title("Draw Here (2D Canvas)", fontsize=12)
    ax3d.set_title("3D Model (Extruded Shape)", fontsize=12)
    fig.canvas.draw()

# Change brush size
def update_brush_size(val):
    global brush_size
    brush_size = int(val)

# Toggle shape mode
def toggle_shape(event):
    global draw_mode
    modes = ["freehand", "circle", "rectangle"]
    draw_mode = modes[(modes.index(draw_mode) + 1) % len(modes)]
    shape_button.label.set_text(f"Mode: {draw_mode.capitalize()}")

# Resize dynamically
def resize_figure(event):
    fig.set_size_inches(get_figure_size(), forward=True)
    fig.canvas.draw()

# Add buttons and slider
ax_clear = plt.axes([0.3, 0.01, 0.1, 0.05])
clear_button = Button(ax_clear, "Clear")
clear_button.on_clicked(clear_canvas)

ax_shape = plt.axes([0.42, 0.01, 0.15, 0.05])
shape_button = Button(ax_shape, "Mode: Freehand")
shape_button.on_clicked(toggle_shape)

ax_slider = plt.axes([0.6, 0.01, 0.3, 0.03])
brush_slider = Slider(ax_slider, "Brush Size", 1, 10, valinit=2)
brush_slider.on_changed(update_brush_size)

# Connect events
fig.canvas.mpl_connect('resize_event', resize_figure)
fig.canvas.mpl_connect('button_press_event', on_press)
fig.canvas.mpl_connect('motion_notify_event', on_motion)
fig.canvas.mpl_connect('button_release_event', on_release)

plt.show()
```
